```python
# Important the dependancies
import pandas as pd
import os
```

```python
# Load in the data
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
len(vine_df.index)
```

```python
# Number of Vine Reviews
vine_reviews = vine_df[vine_df['vine']=='Y']
vine_reviews.info()
```

```python
# Filter the data by votes
filtered = vine_df[vine_df['total_votes'] >=20]
filtered
```

```python
# Filter the data between helpful votes and total votes
votes_df = filtered[filtered['helpful_votes']/filtered['total_votes'] >=.5]
votes_df
```

```python
# Number of Nonvine Reviews
nonvine_reviews = vine_df[vine_df['vine']=='N']
nonvine_reviews.value_counts()
nonvine_reviews
```

```python
# Determining the amount of votes were 'yes'
yes_votes = votes_df[votes_df['vine'] == 'Y']
yes_votes
```

```python
# Grouping the data to determine how many were start_ratings
yes_votes.groupby('star_rating').count()
yes_votes
```

```python
len(yes_votes.index)
```

```python
# Determining the amount of votes that were 'no'
no_votes = votes_df[votes_df['vine'] == 'N'].value_counts()
```

```python
# Grouping the data to determine how many were star_ratings
total_star_ratings = vine_df['star_rating'].count()
total_star_ratings
```

```python
# Total number of reviews 
total_reviews = vine_df.groupby('star_rating').count()
total_reviews
```

```python
# Total amount of all star_ratings and their dtype
vine_df['star_rating'].value_counts('desc')
```

```python
# Total number of reviews 
total_reviews = len(pd.unique(votes_df['review_id']))
print("Total number of reviews:", total_reviews)
```

```python
# Number of paid reviews
paid_reviews = len(votes_df[(votes_df['vine'] == 'Y')])
print("Total number of  paid reviews:", paid_reviews)
```

```python
# Nunber of paid 5 star reviews 
paid_five_star_reviews = len(votes_df[(votes_df['vine'] == 'Y') & (votes_df['star_rating'] ==5)])
print("Total number of paid five star reviews:", paid_five_star_reviews)
```

```python
# Number of unpaid reviews
unpaid_reviews = len(votes_df[(votes_df['vine'] == 'N')])
print("Total number of unpaid reviews:", unpaid_reviews)
```

```python
# The total Number of Unpaid 5-Star Reviews
unpaid_five_star_reviews = votes_df[votes_df['vine']=='N']
unpaid_five_star_reviews = unpaid_five_star_reviews['star_rating'].value_counts()[5]
print("Number of 5 star unpaid reviews:",unpaid_five_star_reviews)
```

```python
# Percentage of Paid 5-Star Reviews 
percent_paid_five_star_reviews = (paid_five_star_reviews/total_reviews)*100
# Round the value 
percent_paid_five_star_rounded = round(percent_paid_five_star_reviews, 6)
print("Percentage of paid five star reviews out of total number reviews:", percent_paid_five_star_rounded)
```

```python
# Percentage of Upaid 5-Star Reviews 
percent_unpaid_five_star_reviews = (unpaid_five_star_reviews/total_reviews)*100

# Round the value 
percent_unpaid_five_star_rounded = round(percent_unpaid_five_star_reviews, 6)
print("Percentage of unpaid five star reviews out of total number reviews:", percent_unpaid_five_star_rounded)
```
